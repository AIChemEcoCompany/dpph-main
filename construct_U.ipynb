```python
import pandas as pd
import numpy as np
from rdkit import Chem
from rdcanon import canon_smarts
from pandarallel import pandarallel
pandarallel.initialize(nb_workers=8,progress_bar=True)
```

```python
df1 = pd.read_csv(r'result/represent_fg1_fg2_1044.csv') #1044 没有考虑去重前 重复
df_1044 = pd.read_csv(r'data/avail_smiles_1044.csv')
```

```python

df1.drop_duplicates('canon_smarts') #1044含有多少键?
```

```python
H_inner_marked_contain_dup = pd.read_csv('data/H_inner_marked_contain_dup.csv',delimiter='\t')
inner_marked_contain_dup = pd.read_csv('data/inner_marked_contain_dup.csv',delimiter='\t')

for index, row in df1.iterrows():
    s = H_inner_marked_contain_dup.loc[H_inner_marked_contain_dup['smarts_inner_marked'] == row['fg1_fg2_marked'],'smarts_marked_oxygen']
    if s.empty:
        continue
    
    v = H_inner_marked_contain_dup.loc[H_inner_marked_contain_dup['smarts_marked_oxygen']==s.values[0], 'smarts_inner_marked'] .values
    for val in v:
        new_row = row.copy()
        new_row['fg1_fg2_marked'] = val
        df1.loc[df1.__len__()] = new_row


for index, row in df1.iterrows():
    s = inner_marked_contain_dup.loc[inner_marked_contain_dup['smarts_marked'] == row['fg1_fg2_marked'],'smarts_marked_oxygen']
    if s.empty:
        continue
    
    v = inner_marked_contain_dup.loc[inner_marked_contain_dup['smarts_marked_oxygen']==s.values[0], 'smarts_marked'] .values
    for val in v:
        new_row = row.copy()
        new_row['fg1_fg2_marked'] = val
        df1.loc[df1.__len__()] = new_row
        
```

```python
from utils.get_marked import get_Hatom1, get_inner_ba12, convert_implicit_H

def deal_H(s:str):
    if ':1' in s:
        mol = Chem.MolFromSmarts(s)
    else:
        s0 = convert_implicit_H(s) #氢
        if s0 == '[N]1(=[C][N;!H0][C][C]1)':
            print(s)
        mol = Chem.MolFromSmarts(s0) 
    return mol

# 1044个化合物有多少个键? 6k 5283

df1['s'] = df1['fg1_fg2_marked'].parallel_apply(deal_H)

```
